## Theme Analysis

```python
plt.figure(figsize=(18,15))
plt.title(f"Theme Distribution (n = {n_lines} | unique = {len(df_lines['theme'].unique())})", size=16)
ax = sns.countplot(data=df_lines, y="theme", edgecolor="black", order=df_lines["theme"].value_counts().index, linewidth=1, palette="Spectral")
ax.bar_label(ax.containers[0])
plt.savefig("../img/barplot_theme.png", dpi=128)
plt.show()

# Get descriptive statistics from text_size grouped by language
df_lines[["text_size", "theme"]].groupby("theme").describe().applymap("{:.2f}".format)
```

## Font Analysis

```python
plt.figure(figsize=(18,10))
plt.title(f"Font Distribution (n = {n_lines} | unique = {len(df_lines['font'].unique())})", size=16)
ax = sns.countplot(data=df_lines, y="font", edgecolor="black", order=df_lines["font"].value_counts().index, linewidth=1, palette="Spectral")
ax.bar_label(ax.containers[0])
plt.savefig("../img/barplot_font.png", dpi=128)
plt.show()

# Get descriptive statistics from text_size grouped by language
df_lines[["text_size", "font"]].groupby("font").describe().applymap("{:.2f}".format)
```

___________________
## Language Analysis
Gives answers about how many different programming languages are used and its distribution.

```python
plt.figure(figsize=(25,5))
plt.title(f"Language Distribution (n = {n_lines} | unique = {len(df_lines['language'].unique())})", size=16)
ax = sns.countplot(data=df_lines, x="language", edgecolor="black", order=df_lines["language"].value_counts().index, linewidth=1, palette="Spectral")
ax.bar_label(ax.containers[0])
plt.savefig("../img/barplot_language.png", dpi=128)
plt.show()

# Get descriptive statistics from text_size grouped by language
df_lines[["text_size", "language"]].groupby("language").describe().applymap("{:.2f}".format)
```

## Target Analysis

```python
plt.figure(figsize=(18,6))
plt.title(f"Analysis of the target size, i.e. text size (Outlier Removed | Threshold = {110} | n = {len(df_lines)} | strip whitespaces)", size=18)
sns.histplot(df_lines.text_size, edgecolor="black")
plt.xticks(np.arange(0, 120, 10))
plt.xlabel("Length of Target", size = 16)
plt.ylabel("Count", size = 16)
plt.savefig("../img/target_analysis.png", dpi=128)
plt.show()
```
